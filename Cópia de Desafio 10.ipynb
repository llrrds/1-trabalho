### Desafio 10

Dada uma lista, divida-a em 3 partes iguais e reverta cada lista.

Exemplo: 

sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]

Parte 1 [8, 45, 11]

Parte 2 [12, 14, 23]

Parte 3 [89, 45, 78]

```python
l = [11, 45, 8, 23, 14, 12, 78, 45, 89]
```

```python
n = 3
splited = []
len_l = len(l)
for i in range(n):
    start = int(i*len_l/n)
    end = int((i+1)*len_l/n)
    splited.append(l[start:end])
print(splited)
```
